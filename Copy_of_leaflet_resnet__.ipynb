```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import VGG19,EfficientNetB3,ResNet50
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D,Flatten,Dropout
```

```python
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True
```

```python

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(1024, activation='relu'),
    Dense(num_classes, activation='softmax')
])
```

```python
from tensorflow.keras.applications.resnet50 import preprocess_input

# Load your CSV file containing image paths and labels
csv_path = '/content/train_shuffle.csv'
csv_valid='/content/validation_file.csv'
df_train = pd.read_csv(csv_path)
df_valid = pd.read_csv(csv_valid)
# Specify the path to your images
train_image_dir = '/content/products_leaflets_256/temp'
val_image_dir='/content/products_leaflets_256/tempvalid'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import preprocess_input
import pandas as pd
import os

# Define parameters for data augmentation and normalization
datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

# Specify batch size and target size for your images
batch_size = 16
target_size = (224, 224)  # adjust based on your model input size

# Create temporary directories to save augmented images for training and validation
train_temp_dir = '/content/train'
val_temp_dir = '/content/validation'
for temp_dir in [train_temp_dir, val_temp_dir]:
    if not os.path.exists(temp_dir):
        os.makedirs(temp_dir)

# Create a generator for training images from the DataFrame
train_generator = datagen.flow_from_dataframe(
    df_train,  # DataFrame for training data
    directory=train_image_dir,
    x_col='Image_Name',
    y_col='Class_Name',
    target_size=target_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    save_to_dir=train_temp_dir,  # Save augmented training images to a temporary directory
    save_prefix='aug_train',     # Prefix for saved images
    save_format='jpeg'           # Save images in JPEG format
)

# Create a generator for validation images from the DataFrame
val_generator = datagen.flow_from_dataframe(
    df_valid,  # DataFrame for validation data
    directory=val_image_dir,
    x_col='Image_Name',
    y_col='Class_Name',
    target_size=target_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False,  # No need to shuffle validation data
    save_to_dir=val_temp_dir,  # Save augmented validation images to a temporary directory
    save_prefix='aug_val',     # Prefix for saved images
    save_format='jpeg'          # Save images in JPEG format
)
```

```python
sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.95)
model.compile(optimizer=sgd_optimizer ,loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=15,
    validation_data=val_generator,
    validation_steps=val_generator.samples // batch_size
)


```

```python
import pandas as pd
import os
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np

# Path to the CSV file containing test image names and labels
csv_file_path = '/content/test_class_mapping.csv'

# Path to the directory containing the test images
test_images_directory = '/content/products_leaflets_256/test_temp'

# Load CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Load pre-trained ResNet50 model (assuming you have already loaded and assigned it to the 'loaded_model' variable)
# Replace the following line with the actual loading of your model
# Load your pre-trained model

# Create a new column in the DataFrame to store predictions
df['Predictions'] = np.nan

# Function to preprocess and predict an image
def preprocess_and_predict(image_path):
    # Load the image
    img = image.load_img(image_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # Predict the class probabilities
    predictions = loaded_model.predict(x)

    # Decode and get the class with the highest probability
    predicted_class = np.argmax(predictions)

    return predicted_class

# Loop through each row in the DataFrame
for index, row in df.iterrows():
    image_name = row['Image_Name']
    image_path = os.path.join(test_images_directory, image_name)

    # Ensure the image file exists
    if os.path.exists(image_path):
        # Preprocess and predict the image
        predicted_class = preprocess_and_predict(image_path)

        # Assign the predicted class to the 'Predictions' column in the DataFrame
        df.at[index, 'Predictions'] = predicted_class
    else:
        print("Image file not found: {}".format(image_path))

# Save the updated DataFrame to a new CSV file
output_csv_path = '/content/predictions.csv'
df.to_csv(output_csv_path, index=False)
print(f"Predictions saved to: {output_csv_path}")
```
